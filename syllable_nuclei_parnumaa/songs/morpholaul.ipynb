```python
import nltk 
import estnltk
from estnltk import Text 
from estnltk.taggers import VabamorfTagger
from estnltk.default_resolver import make_resolver
# Create morphological tagger with default settings




morph_tagger = VabamorfTagger(disambiguate=False, phonetic=True)
morph_tagger

laul = Text('Luud mina võtin lutsu suusta, laulud lattika nenasta, havvi hammaste vahelta,luhas loogu võta dessa, ja arus äästa dessa.')
laul.text
```

```python
laul.tag_layer(['words','sentences','compound_tokens'])
```

```python
morph_tagger(laul)
```
